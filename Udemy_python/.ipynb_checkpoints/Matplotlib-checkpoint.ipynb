```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,100,100)
y = x**3
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.set_xticks([])
axes1.set_yticks([])
axes2.set_xticks([])
axes2.set_yticks([])

axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title('LARGER PLOT')
axes2.set_title('SMALLER PLOT')
```

```python
fig ,axes = plt.subplots(nrows=3 , ncols=3)

plt.tight_layout()
```

```python
fig ,axes = plt.subplots(nrows=1 , ncols=2)

axes[0].plot(x,y)
axes[0].set_title('FIRST PLOT')

axes[1].plot(y,x)
axes[1].set_title('SECOND PLOT')
```

```python
fig ,axes = plt.subplots(nrows=1 , ncols=2 , figsize=(14,8))

axes[0].plot(x[:11],np.array(x**2)[:11] , label='X squared' , linestyle='--' , marker='s',\
            markersize=10, markerfacecolor='yellow' , markeredgecolor='red' , markeredgewidth=2)
axes[0].plot(x[:11],np.array(x**3)[:11] , label='X cubed'  ,color='orange' ,ls = '-',lw=1 , marker='o' , markersize=20 ,\
            markerfacecolor='yellow' , markeredgewidth=3 , markeredgecolor='red') #ls=linestyle
axes[0].legend(loc = 0)
axes[0].set_title('FIRST PLOT')

axes[1].plot(x,x**2 , label='X squared',color='blue' , linewidth=2) #default linewidth is equal to 1
axes[1].plot(x,x**3 , label='X cubed' , color='#FF8C00',lw=4 ) #RGB Hex code #lw = linewidth
axes[1].legend(loc=(0.2,0.8))
axes[1].set_title('SECOND PLOT')

sub_ax = fig.add_axes([0.17,0.55,0.2,0.2])
# sub_ax.set_xticks([])
# sub_ax.set_yticks([])
sub_ax.plot(x[:11],np.array(x**3)[:11] , label='X cubed'  ,color='orange' ,ls = '-',lw=1 , marker='o' , markersize=10 ,\
            markerfacecolor='yellow' , markeredgewidth=3/5 , markeredgecolor='red')
sub_ax.set_xlim([2,6])
sub_ax.set_ylim([2**3,6**3])
```
