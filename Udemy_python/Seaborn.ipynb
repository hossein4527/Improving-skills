```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips[tips.columns[0]] , kde=True\
             , bins=30, )
```

```python
sns.jointplot(x = tips.columns[0],y=tips.columns[1] , data=tips, \
             kind='scatter')
```

```python
sns.pairplot(data=tips , hue='sex',)
```

```python
sns.rugplot(tips[tips.columns[0]])
```

```python
sns.distplot(tips[tips.columns[0]], kde=True , bins=30)
```

```python
sns.kdeplot(tips[tips.columns[0]])
```

```python
sns.barplot(x='sex' , y='total_bill',data = tips\
           , estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day' , y='total_bill', data=tips, \
           hue='smoker')
```

```python
sns.violinplot(x='day' , y='total_bill', data=tips, \
           hue='sex' , split=True)
```

```python
sns.stripplot(x = 'day' , y='total_bill' , data=tips\
             ,jitter=True , hue='sex' , split=True)
```

# Grids

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris[iris.columns[-1]].unique()
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
h  = sns.FacetGrid(data=tips , col='time',row='smoker')
h.map(sns.distplot , 'total_bill')
```

```python
h  = sns.FacetGrid(data=tips , col='time',row='smoker')
h.map(plt.scatter , 'total_bill' , 'tip')
```

# Regression


```python
sns.set_style("darkgrid")
sns.lmplot(x='total_bill' , y='tip' ,data=tips,\
          hue='sex', markers=['o','v'], \
          scatter_kws={'s':40})
```

```python
sns.set_style("darkgrid")
sns.lmplot(x='total_bill' , y='tip' ,data=tips,\
          col='day',hue='sex', \
          scatter_kws={'s':40},\
          aspect=0.6, size=8)
```

# Style and Color

```python
plt.figure(figsize=(6,4))
sns.set_style('ticks')
sns.countplot(x='sex' , data=tips)
# sns.despine(left=True, bottom=True)
```

```python
plt.figure(figsize=(6,4))
sns.set_context('poster' , font_scale=1)
sns.countplot(x='sex' , data=tips)
# sns.despine(left=True, bottom=True)
```

```python
sns.set_context("notebook")
sns.lmplot(x='total_bill' , y='tip' ,data=tips,\
          hue='sex', markers=['o','v'], \
          scatter_kws={'s':40, 'alpha':0.8}, palette='seismic')
```
